```python
import pandas as pd
import os
import re
import nltk

#nltk.download('punkt')
```

```python
got = pd.read_csv("Game_of_Thrones_Script/Game_of_Thrones_Script.csv")
got.name = 'Game of Thrones'

got
```

```python
df = got[["Name","Sentence"]]

df = df.rename(columns = {"Name":"character","Sentence":"line"})
df
```

```python
from nltk import word_tokenize 
from nltk.util import ngrams

def get_ngrams(ngram):
    data = list()
    counter = 0
    for line in df['line']:
        token = nltk.word_tokenize(line)
        bigram = list(ngrams(token, ngram))
        data.append([df['character'][counter],bigram])
        counter = counter + 1
    return data
        
data = get_ngrams(2)
```

```python
newdf = pd.DataFrame(data, columns=['character','bigram'])
newdf
```
